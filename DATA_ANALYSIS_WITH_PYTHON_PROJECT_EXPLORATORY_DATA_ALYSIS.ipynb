```python
Title=[]
Price=[]
Updated=[]
Owner=[]
Model=[]
Location=[]
kms=[]
for i in range(1,30):
    url='https://www.bikewale.com/used/bikes-in-india/'.format(i)
    #print (url)
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    a=soup.find_all('a',class_="text-truncate text-black")
    for i in a:
        Title.append(i.text)
 
    #to get the price
    
    for b in soup.find_all('span',class_="font22 text-bold"):
          Price.append(b.text) 

    #to get the updated
    
    for c in soup.find_all("span",class_="font12 text-xt-light-grey"):
          Updated.append(c.text)
            
    d=soup.find_all('div',class_="grid-6 alpha")
    
     # to get owner and model
        
    for i in range(0,len(d)):
        if(i%2==1):
            Owner.append(d[i].text)
        else:
            Model.append(d[i].text)
        
    # to get the location and the kms 
        
    e=soup.find_all('div',class_="grid-6 alpha omega")
    for i in range(0,len(e)):
        if(i%2==1):
            Location.append(e[i].text)
        else:
            kms.append(e[i].text)
```

```python
print(len(Title))
print(len(Price))
print(len(Updated))
print(len(Owner))
print(len(Model))
print(len(Location))
print(len(kms))
```

```python
year_m=[]
Title=[]
for i in Bikes['Title']:
    a=i.split(',')   # year split
    year_m.append(a[0])
    Title.append(a[1])
```

```python
Bikes=pd.DataFrame({'Title':Title, 'Price':Price, 'Updated':Updated, 'Owner':Owner, 'Model':Model, 'Location':Location, 'Kms':kms})
```

```python
year_m=[]
brand=[]
for i in Bikes['Title']:
    b=i.split(' ')
    year_m.append(b[0])
    brand.append(b[1:2])
```

```python
brand
```

```python
len(brand)
```

```python
Bikes=pd.DataFrame({'Brand':brand,'Title':Title, 'Price':Price, 'Updated':Updated, 'Owner':Owner, 'Model':Model, 'Location':Location, 'Kms':kms})
```

```python
Bikes
```

```python
m=[]
for i in Bikes['Brand']:
    a=str(i).strip('[]')
    b=a.split(',')
    m.append(''.join(b))
Bikes['Brand']=m
```

```python
Bikes
```

```python
r=[]
for i in Bikes['Brand']:
    r.append(i.strip("''"))
Bikes['Brand']=r
```

```python
Bikes
```

```python
h=[]
for i in Bikes['Price']:
    a=i.split(',')
    h.append(''.join(a))
Bikes['Price']=h
```

```python
Bikes
```

```python
regex=r':(.*)'

Bikes['Updated']=Bikes['Updated'].apply(lambda x:re.findall(regex,x))
```

```python
Bikes
```

```python
Bikes['Updated']=Bikes['Updated'].apply(lambda x:''.join(x))
```

```python
Bikes
```

```python
Bikes['Owner'].apply(lambda x:''.join(x))
```

```python
Bikes['Owner']=Bikes['Owner'].str.replace('Owner','')
```

```python
Bikes
```

```python
 Bikes['Model'].apply(lambda x:''.join(x))
```

```python
Bikes['Model']=Bikes['Model'].str.replace('model','')
```

```python
Bikes
```

```python
j=[]
for i in Bikes['Kms']:
    a=str(i).split('kms')
    j.append(''.join(a))
Bikes['Kms']=j
```
